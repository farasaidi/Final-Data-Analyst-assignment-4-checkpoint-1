## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# # Load the data into a Pandas DataFrame
# url = "https://www.kaggle.com/datasets/imakash3011/rental-bike-sharing"
# df1 = pd.read_csv(url)
```

```python
# Explore the data with some basic statistics
print(df.describe())
```

```python
# Plot a histogram of the bike rental counts
sns.histplot(data=df, x="cnt", kde=True)
plt.title("Distribution of Bike Rental Counts")
plt.xlabel("Bike Rental Count")
plt.ylabel("Frequency")
plt.show()

```

```python
# Plot a scatter plot of temperature vs bike rental count
sns.scatterplot(data=df, x="temp", y="cnt")
plt.title("Bike Rental Count vs Temperature")
plt.xlabel("Temperature (Celsius)")
plt.ylabel("Bike Rental Count")
plt.show()
```

```python
# Plot a bar chart of the average bike rental count by season
season_counts = df.groupby("season")["cnt"].mean()
sns.barplot(x=season_counts.index, y=season_counts.values)
plt.title("Average Bike Rental Count by Season")
plt.xlabel("Season")
plt.ylabel("Average Bike Rental Count")
plt.show()
```
